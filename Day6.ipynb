```python
'''
0 1 2 2 3 5 8 13 

fib(0)=0
fib(1)=1    
fib(2)=fib(0)+fib(1)=0+1=1
fib(3)=fib(1)+fib(2)=1+1=2
fib(4)=fib(2)+fib(3)=1+2=3
fib(n)=fib(n-2)+fib(n-1)

'''
def fib(n):
    #base case of recursion
    if(n==0 or n==1):
        return n
    #recursive case
    return fib(n-2)+fib(n-1)
print(fib(0))
print(fib(1))
print(fib(2))
print(fib(3))
print(fib(4))
print(fib(5))   

    
```

```python
x = 10  # Global variable

def my_func():
    x = 5  # Local variable
    print(x)  # Output: 5

my_func()
print(x)  # Output: 10 (global x remains unchanged)
```

```python
x = 60  # Global variable

def modify_global():
    global x
    x = 9 # Modifies the global x

modify_global()
print(x)  # Output: 5
```
