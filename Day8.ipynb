```python
A = {1, 2, 3, 4}
print(A)
```

```python
#if has Duplicate, it will be removed
A = {1, 2, 2, 3, 4, 4}
print(A)
```

```python
#Union:dono sets ke saare unique elements
A = {1,2,3}
B = {3,4,5}

print(A | B)
```

```python
A.union(B)
```

```python
#Intersection:common elements
A = {1,2,3}
B = {2,3,4}

print(A & B)
```

```python
A.intersection(B)
```

```python
#Difference: A me se B ke elements remove kr dega
A = {1,2,3}
B = {2,3,4}

print(A - B)
```

```python
#Symmetric Difference: dono sets ke aise elements jo common nhi hai
A = {1,2,3}
B = {3,4,5}

print(A ^ B)
```

```python
#Subset: A ke sare elements B me hai ya nhi
A = {1,2}
B = {1,2,3,4}

print(A.issubset(B))
```

```python
#Superset: B ke sare elements A me hai ya nhi
B.issuperset(A)
```

```python
A = {1,2,3}

A.add(4)

print(A)
```

```python
A.remove(2)
print(A)
```

```python
# Empty dictionary
empty_dict = {}

# Dictionary with key-value pairs
student = {
    "name": "Alice",
    "age": 25,
    "grade": "A"
}

# Using dict() constructor
person = dict(name="John", age=30, city="New York")
```

```python
# Using keys
print(student["name"])  # Alice

# Using get() (avoids KeyError if key doesn't exist)
print(student.get("age"))  # 25
print(student.get("height", "Not Found"))  # Default value
```

```python
student = {"name": "Alice", "age": 25, "grade": "A"}

# Adding a new key-value pair
student["city"] = "New York"

# Updating an existing value
student["age"] = 26

# Removing an item
student.pop("grade")

# Iterating over a dictionary
for key, value in student.items():
    print(key, ":", value)

# Output:
# name : Alice
# age : 26
# city : New York
```

```python
# Creating a dictionary using comprehension
squares = {x: x**2 for x in range(1, 6)}
print(squares)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```
