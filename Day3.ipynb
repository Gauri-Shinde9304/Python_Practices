```python
'''Age: 10, 18, 25, 16
👉 Naya column "Status":

Age ≥ 18 → "Adult"

Else → "Minor"

'''
ages = [10, 18, 25, 16]
df=pd.DataFrame(ages, columns=['Age'])
df['Status'] = df['Age'].apply(lambda x: 'Adult' if x >= 18 else 'Minor')
print(df)
```

```python
'''Values: -5, 0, 7, -2
👉 Naya column "Nature":

0 → "Positive"

= 0 → "Zero"

< 0 → "Negative"'''
values = [-5, 0, 7, -2]
df =pd.DataFrame(values, columns=['Values'])
df['Nature']= df['Values'].apply(lambda x : 'Positive' if x>0 else ('Zero' if x==0 else 'Negative'))
print(df)
```

```python
'''Data:

Marks: 92, 67, 45, 30


👉 Column "Grade":

≥ 90 → A

≥ 60 → B

≥ 40 → C

Else → Fail'''
marks =[92, 67, 45, 30, 57, 22, 45]
df = pd.DataFrame(marks, columns=['Marks'])
df['Grade']=df['Marks'].apply(lambda x: 'A'if x>=90 else ('B' if x>=60 else ('C' if x>=40 else 'fail')))
print(df)
```

```python
'''Salary: 20000, 45000, 80000
👉 Column "Tax":

< 30000 → "No Tax"

30000–60000 → "10% Tax"

60000 → "20% Tax"'''
salaries = [20000, 45000, 80000]
df = pd.DataFrame(salaries, columns=['Salary'])
df['Tax'] = df['Salary'].apply(lambda x: 'No Tax' if x<30000 else ('10% Tax' if x<=60000 else '20% Tax'))
print(df)
```

```python
# Tax Calculation Based on Salary and Tax Percent
import pandas as pd

def calculate_tax(salary, tax_percent):
    tax_amt = salary * tax_percent / 100 
    net_salary = salary - tax_amt             
    return tax_amt, net_salary

salary = float(input("Enter your salary: "))
tax_percent = float(input("Enter tax percent: "))

tax_amt, net_salary = calculate_tax(salary, tax_percent)

df = pd.DataFrame({
    'Salary': [salary],
    'Tax Percent': [tax_percent],
    'Tax Amount': [tax_amt],
    'Net Salary': [net_salary]
})

print(df)
```

```python
# Tax Calculation for Multiple Salaries and Tax Percentages
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Salary": [20000, 45000, 80000],
    "Tax %": [0, 10, 20]
})

df["Tax Amount"] = np.where(
    df["Tax %"] == 0, 0,
    df["Salary"] * (df["Tax %"] / 100)
)

df["Net Salary"] = df["Salary"] - df["Tax Amount"]

print(df)

```

```python
# Tax Calculation for Multiple Salaries and Tax Percentages
import pandas as pd

data = {
    "Salary": [20000, 45000, 80000],
    "Tax %": [0, 10, 20]
}

df = pd.DataFrame(data)

df["Tax Amount"] = df.apply(
    lambda row: row["Salary"] * (row["Tax %"] / 100),
    axis=1
)

df["Net Salary"] = df["Salary"] - df["Tax Amount"]

print(df)
```

```python
#Dictionary Mapping Approach
import pandas as pd

salary = float(input("Enter Salary: "))
tax_percent = float(input("Enter Tax Percentage: "))

tax_rates = {
    0: 0,
    10: 0.10,
    20: 0.20
}

tax_rate = tax_rates.get(tax_percent, tax_percent / 100)

tax_amount = salary * tax_rate
net_salary = salary - tax_amount

df = pd.DataFrame({
    "Salary": [salary],
    "Tax %": [tax_percent],
    "Tax Amount": [tax_amount],
    "Net Salary": [net_salary]
})

print(df)
```
