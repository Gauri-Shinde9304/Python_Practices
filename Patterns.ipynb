```python
n = 5

for i in range(n):                 # 0,1,2,3,4
    for s in range(n - i - 1):     # spaces
        print(" ", end=" ")
    for j in range(i + 1):         # stars
        print("*", end=" ")
    print()
```

```python
n = 5
for i in range(n):

    for j in range(i + 1):
        print(' ', end='')

    for j in range(i, n):
        print('*', end=' ')

    print()
```

```python
n = 5

for i in range(n):
    # spaces
    for j in range(n - i - 1):
        print(' ', end='')

    # stars
    for j in range(2 * i + 1):
        print('*', end='')

    print()
```
