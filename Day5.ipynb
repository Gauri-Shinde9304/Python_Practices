```python
#Concatenate two strings: "Hello" and "World" with a space in between.
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2
print(result)
```

```python
'''Given text = "Python Programming", do the following:

Print the first 6 characters
Print the last 6 characters
Print every second character from the string'''
text = "Python Programming"
print(text[:6])          # First 6 characters
print(text[-6:])         # Last 6 characters
print(text[::2])        # Every second character    
```

```python
#Reverse the string text using slicing.
reversed_text = text[::-1]
print(reversed_text)
```

```python
'''Take the string "  i love python programming  " and:

Remove extra spaces from both ends
Convert it to title case
Count how many times "o" appears'''
sample_str = "  i love python programming  "
trimmed_str = sample_str.strip()
title_case_str = trimmed_str.title()
o_count = trimmed_str.count('o')
print(f"Trimmed: '{trimmed_str}'")
print(f"Title Case: '{title_case_str}'")
print(f"Count of 'o': {o_count}")
    
```

```python
# if the string "123abc" is alphanumeric.
test_str = "123abc"
is_alphanumeric = test_str.isalnum()
print(f"Is '{test_str}' alphanumeric? {is_alphanumeric}")
```

```python
'''Using format(), create a sentence:
"My name is John and I am 25 years old."
by passing "John" and 25 as variables.'''
name = input("Enter Your Name")
age = input("Enter Your Age")
sentence = "My name is {} and I am {} years old.".format(name, age)
print(sentence) 
```

```python
'''Using format(), create a sentence:
"My name is John and I am 25 years old."
by passing "John" and 25 as variables name should not be more than 10 characters and age should be a 2 digit number.'''
name = input("Enter Your Name (max 10 characters): ")
if len(name) > 10:
    print("Invalid name. Name must be maximum 10 characters.")
else:
    age = input("Enter Your Age (2 digit number): ")
    if not age.isdigit():
        print("Invalid age. Age must be numeric.")
    elif int(age) < 10 or int(age) > 99:
        print("Invalid age. Age must be a 2 digit number.")
    else:
        sentence = "My name is {} and I am {} years old.".format(name, age)
        print(sentence)
```
