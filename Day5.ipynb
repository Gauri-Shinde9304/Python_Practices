```python
'''Take the string "  i love python programming  " and:

Remove extra spaces from both ends
Convert it to title case
Count how many times "o" appears'''
sample_str = "  i love python programming  "
trimmed_str = sample_str.strip()
title_case_str = trimmed_str.title()
o_count = trimmed_str.count('o')
print(f"Trimmed: '{trimmed_str}'")
print(f"Title Case: '{title_case_str}'")
print(f"Count of 'o': {o_count}")
    
```

```python
'''Using format(), create a sentence:
"My name is John and I am 25 years old."
by passing "John" and 25 as variables name should not be more than 10 characters and age should be a 2 digit number.'''
name = input("Enter Your Name (max 10 characters): ")
if len(name) > 10:
    print("Invalid name. Name must be maximum 10 characters.")
else:
    age = input("Enter Your Age (2 digit number): ")
    if not age.isdigit():
        print("Invalid age. Age must be numeric.")
    elif int(age) < 10 or int(age) > 99:
        print("Invalid age. Age must be a 2 digit number.")
    else:
        sentence = "My name is {} and I am {} years old.".format(name, age)
        print(sentence)
```

```python
'''Write a program that counts how many vowels are in a given string.'''
input_string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for char in input_string:
    if char in vowels:
        count += 1
print(f"Number of vowels: {count}")

```

```python
'''Take a user input string and check if it is a palindrome (same forwards and backwards).'''
input_string = input("Enter a string: ")
reversed_string = input_string[::-1]
if input_string == reversed_string:
    print(f"'{input_string}' is a palindrome.")
```

```python
import pandas as pd
data = {
    "Animals": [
        "rat", "rabbit", "dog", "rider", "cat",
        "ram", "lion", "tiger", "rhino", "rat"
    ],
    "Fruits": [
        "apple", "banana", "raspberry", "grape", "redberry",
        "mango", "roseapple", "orange", "raspberry", "kiwi"
    ],
    "TreeNames": [
        "rosewood", "neem", "peepal", "rubber", "oak",
        "rattan", "maple", "redwood", "teak", "rosewood"
    ]
}
df = pd.DataFrame(data)
# Convert to long format
df = df.melt(var_name="Category", value_name="Word")
# Add row numbers
df["Row"] = df.groupby("Category").cumcount() + 1
# Keep only words starting with 'r'
df = df[df["Word"].str.startswith("r")].copy()
# Count repetitions inside each category
df["Repeated"] = df.groupby(["Category", "Word"])["Word"].transform("count")
print(df)
```
