```python
import json
```

```python

data = {"name": "Alice", "age": 25}

# dumps: convert dict to json string
json_string = json.dumps(data)

# loads: convert json string back to dict
dict_from_string = json.loads(json_string)

# dump (write to file)
with open("data.json", "w") as f:
    json.dump(data, f)

# load (read from file)
with open("data.json", "r") as f:
    dict_from_file = json.load(f)


print("| Function  | Output |")
print("|-----------|--------|")
print(f"| dumps() | {json_string} |")
print(f"| loads() | {dict_from_string} |")
print()
print("| Function  | Output |")
print("|-----------|--------|")
print(f"| dump() | data saved to file 'data.json' |")
print()
print("| Function  | Output |")
print("|-----------|--------|")
print(f"| load() | {dict_from_file} |")
```

```python
from abc import ABC, abstractmethod

# 🔹 ABSTRACTION
class Person(ABC):
    def __init__(self, name, age):
        self._name = name      # 🔒 Encapsulation
        self._age = age

    @abstractmethod
    def get_role(self):
        pass


# 🔹 INHERITANCE
class Student(Person):
    def __init__(self, name, age, marks):
        super().__init__(name, age)
        self.marks = marks

    def get_role(self):
        return "Student"

    def display(self):
        print(f"Name: {self._name}, Age: {self._age}, Marks: {self.marks}")


class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def get_role(self):
        return "Teacher"

    def display(self):
        print(f"Name: {self._name}, Age: {self._age}, Subject: {self.subject}")


# 🔹 POLYMORPHISM
def show_role(person):
    print("Role:", person.get_role())


# 🔹 OBJECTS
s1 = Student("Gauri", 20, 85)
t1 = Teacher("Wahid", 40, "Math")

s1.display()
t1.display()

show_role(s1)
show_role(t1)
```
