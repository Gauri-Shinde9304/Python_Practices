```python
empty_list=[] #or  #list[]
my_list=['apple' ,'banana','orange','grape','kiwi','pear','peach','plum','mango','melon']
print(my_list)
print(my_list[0]) #first item
print(my_list[-3]) #third item from last
print(my_list[3:5]) #slicing from index 1 to 3
print(my_list[2:]) #slicing from index 2 to end
print(my_list[::2]) #slicing with step 2
print(my_list[:6]) #slicing from start to index 5
print(my_list[2:8:3]) #slicing from index 2 to 7 with step 3
print(my_list[::-1]) #reversing the list
```

```python
# List operation
my_list.append('raspberry')#add item at the end
my_list.insert(2,'cranberry')#add item at index 2
my_list.remove('banana') #remove item by value
my_list.pop()               # last item remove karta hai
my_list.pop(3)              # index 3 ka item remove karta hai
my_list.extend(['fig','date'])  # ek se zyada items add karta hai
#my_list.clear()             # puri list empty kar deta hai
print(my_list.index('kiwi'))   # 'kiwi' ka index batata hai
print(my_list.count('apple')) # 'apple' kitni baar hai
my_list.sort()              # list ko A-Z order me sort karta hai
my_list.reverse()           # list ko reverse karta hai
new_list = my_list.copy()   # list ka copy banata hai
print('apple' in my_list)       # True/False check
print('berry' not in my_list)
print(len(my_list))         # list ka size batata hai
print(sorted(my_list))
print(my_list)
```
